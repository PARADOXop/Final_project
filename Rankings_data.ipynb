```python
import pandas as pd

rankings_df = pd.read_csv("rankings_clean(1).csv")
competitors_df = pd.read_csv("competitors_clean(1).csv")

print("Loaded files successfully")
```

```python
rankings_df = rankings_df.drop_duplicates()
competitors_df = competitors_df.drop_duplicates()

print("Duplicates removed")
```

```python
print(rankings_df.isnull().sum())
print(competitors_df.isnull().sum())
```

```python
rankings_df = rankings_df.fillna({
    "movement": 0,
    "competitions_played": 0
})

competitors_df = competitors_df.fillna("Unknown")

print("Null values handled")
```

```python
rankings_df["rank"] = rankings_df["rank"].astype(int)
rankings_df["points"] = rankings_df["points"].astype(int)
rankings_df["movement"] = rankings_df["movement"].astype(int)
rankings_df["competitions_played"] = rankings_df["competitions_played"].astype(int)

print("Data types assigned")
```

```python
rankings_df = rankings_df.rename(columns={
    "competitor_id": "player_id"
})

competitors_df = competitors_df.rename(columns={
    "competitor_id": "player_id"
})

print("Columns renamed")
```

```python
rankings_df["rank_category"] = rankings_df["rank"].apply(
    lambda x: "Top 10" if x <= 10 else "Top 100" if x <= 100 else "Other"
)
```

```python
assert (rankings_df["points"] >= 0).all()
assert (rankings_df["rank"] > 0).all()
```

```python
print(rankings_df[["rank", "points", "competitions_played"]].describe())
```

```python
rankings_df.to_csv("rankings_final.csv", index=False)
competitors_df.to_csv("competitors_final.csv", index=False)

print("Final cleaned files saved")
```
